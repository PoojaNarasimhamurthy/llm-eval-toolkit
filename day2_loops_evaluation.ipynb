```python
# AI model evaluation loop
models = ["GPT-4", "Claude-3", "Gemini", "Llama"]
scores = [0.92, 0.88, 0.95, 0.79]
threshold = 0.90

print(f"{'='*35}")
print(f"  MODEL EVALUATION RESULTS")
print(f"{'='*35}")

for i, model in enumerate(models):
    result = "PASS" if scores[i] >= threshold else "FAIL"
    print(f"  {model:<12}: {scores[i]:.2%} — {result}")

print(f"{'='*35}")
```
